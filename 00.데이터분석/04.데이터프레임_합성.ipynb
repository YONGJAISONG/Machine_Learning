```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
 '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
 '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])

df1
```

```python
df2 = pd.DataFrame({
 '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
 '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])

df2
```

```python
# Inner Join
pd.merge(df1, df2)
```

```python
# Left Outer Join
pd.merge(df1, df2, how='left')
```

```python
# Right Outer Join
pd.merge(df1, df2, how='right')
```

```python
# Full Outer Join
pd.merge(df1, df2, how='outer')
```

* Join method

```python
df2.join(df1)
```

```python
df3 = df1.set_index('고객번호')
df4 = df2.set_index('고객번호')
df3.join(df4)
```
