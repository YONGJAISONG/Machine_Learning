```python
# 함수에는 series가 들어가야 한다.

def get_newstats(s):
    Q1 = s.quantile(.25)
    Q3 = s.quantile(.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q3 + 1.5*IQR

    for i in s.index:
        s[s[i] > upper_limit | s[i] < lower_limit] = np.nan 

    return s.mean(), s.std()
```
