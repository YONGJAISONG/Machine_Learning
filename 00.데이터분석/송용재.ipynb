# 선생님 해설

### 1. cctv데이터

```python
cctv = pd.read_csv('서울시CCTV설치운영현황(자치구)_년도별_210731기준.csv', skiprows = 1, encoding = 'EUC-KR')
cctv.head(3)
```

```python
# Null data가 있는지 확인
cctv.isna().sum().sum()
```

```python
# 1000단위 구분기호 없애기
for column in cctv.columns[1:]:
    cctv[column] = cctv[column].apply(lambda x : int(x.replace(",", "")))
cctv.head()
```

```python
# 최근 증가율
# 참고로 컬럼 새로 추가할 때는 cctv['컬럼명']의 형태로 써줘야 한다.

cctv['최근 3개년'] = cctv['2018년'] + cctv['2019년'] + cctv['2020년']
cctv['2017년 이전'] = cctv['총계'] - cctv['최근 3개년'] - cctv['2021년']
cctv['최근증가율'] = round(cctv['최근 3개년'] / cctv['2017년 이전'] * 100, 2)
cctv.head(3)
```

```python
cctv['최근증가율2'] = round(cctv.iloc[:, 9:12].sum(axis = 1) / cctv.iloc[:, 2:9].sum(axis = 1)*100, 2)
cctv.head()
```

```python
# Selection - 필요한 컬럼 추출
cctv = cctv[['구분', '총계', '최근증가율']]
cctv.head()
```

```python
# 컬럼명 변경
cctv.rename(columns = {'구분' : '구별',
                       '총계' : 'CCTV댓수'}, inplace=True)
cctv.head()
```

```python
# 첫번째 행(계) 삭제
cctv.drop([0], inplace = True)
```

### 2. 인구 데이터

```python
pop = pd.read_csv('서울시주민등록인구(구별)통계.txt', skiprows = 2, sep = '\t', encoding = 'utf-8')
pop.head()
```

```python
# Null 데이터 확인
pop.isna().sum().sum()
```

```python
# 필요한 컬럼 추출 및 컬럼명 변경
pop = pop[['자치구', '계', '계.2', '65세이상고령자']]
pop.columns = ['구별', '인구수', '외국인', '고령자']
pop.head()
```

```python
# 1000단위 구분기호 없애기
for column in pop.columns[1:]:
    pop[column] = pop[column].str.replace(',', '').astype(int)

pop.head()
```

```python
# 첫번째 행 지우기

pop.drop([0], inplace=True)
```

```python
# 외국인비율, 고령자비율 컬럼 만들기
pop['외국인비율'] = round(pop['외국인'] / pop['인구수'] *100, 2)
pop['고령자비율'] = round(pop['고령자'] / pop['인구수'] *100, 2)
pop.head()
```

### 3. 두 개의 데이터프레임 합치기

```python
df = pd.merge(cctv, pop)
df.head()
```

```python
df.set_index('구별', inplace = True)
df.head()
```

```python
df.to_csv('서울시_CCTV_인구_병합데이터.csv')
```
