# 다중회귀 - 보스턴 주택 가격

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

* 산점도와 선형회귀선

```python
sns.regplot(x = 'CRIM', y = 'PRICE', data = df)
plt.show()
```

```python
_, axs = plt.subplots(nrows=3, ncols=4, figsize=(16, 12))
_, axs
```

```python
_, axs = plt.subplots(nrows=3, ncols=4, figsize=(16, 12))
for i, feature in enumerate(df.columns[1:-1]):
    row, col = i//4, i%4
    sns.regplot(x=feature, y='PRICE', data=df, ax = axs[row][col])
```

* feature별 R squared value

```python
from sklearn.linear_model import LinearRegression
```

```python
# CRIM
X = df.CRIM.values.reshape(-1, 1)
y = boston.target
lr = LinearRegression()
lr.fit(X, y)
lr.score(X, y)
```

```python
y = boston.target
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1, 1)
    lr = LinearRegression()
    lr.fit(X, y)
    score = lr.score(X, y)
    print(f'{feature}:\t{score : .4f}')
```

* 다중 선형회귀

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size = 0.1, random_state=2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 다중 선형회귀에서는 변수가 늘어나면 R^2이 증가하므로 R squared value가 무의미

lr.score(X_test, y_test)
```

```python
# Weight
lr.coef_
```

```python
lr.intercept_
```

* 회귀식

PRICE = -0.01*CRIM + 0.053*ZN + ... - 0.55*LSTAT + 35.5


* 테스트 데이터

```python
X_test[0], y_test[0]
```

```python
# 예측값

np.dot(X_test[0], lr.coef_) + lr.intercept_
```

```python
# predict로 예측한 값
lr.predict(X_test[0].reshape(1, -1))[0]

# 참고) 슬라이싱할 경우 데이터가 2차원으로 나온다
# X_test[:1]
```

```python
# 10개의 데이터에 대한 예측값과 실제값
for i in range(10):
    pred1 = np.dot(X_test[i], lr.coef_) + lr.intercept_
    # pred2 = lr.predict(X_test[i].reshape(1, -1))
    pred2 = lr.predict([X_test[i]])
    # pred2 = lr.predict(X_test[i:i+1])
    print(f'실제값 : {y_test[i]}, 예측값1 : {pred1:.2f}, 예측값2 : {pred2[0]:.2f}')
```
